```python
import matplotlib.pyplot as plt
import numpy as np

# Data
classifiers = ['PAC', 'SGD', 'Hoeffding Tree', 'Gradient Boosting']
accuracy = [0.9935, 0.9787, 0.9918, 0.9953]
precision = [0.9935, 0.9854, 0.9987, 0.9946]
recall = [0.9933, 0.9744, 0.9858, 0.9966]
f1_score = [0.9934, 0.9799, 0.9922, 0.9956]

# Set the width of the bars
bar_width = 0.2
index = np.arange(len(classifiers))

# Create grouped bar graph
plt.figure(figsize=(12, 8))  # Adjust figure size
plt.bar(index - bar_width * 1.5, accuracy, bar_width, label='Accuracy', color='skyblue')
plt.bar(index - bar_width / 2, precision, bar_width, label='Precision', color='lightgreen')
plt.bar(index + bar_width / 2, recall, bar_width, label='Recall', color='orange')
plt.bar(index + bar_width * 1.5, f1_score, bar_width, label='F1-score', color='lightcoral')

# Add labels and title
plt.xlabel('Classifiers', fontsize=14)  # Adjust font size
plt.ylabel('Metrics', fontsize=14)  # Adjust font size
plt.title('Performance Metrics of Classifiers', fontsize=16)  # Adjust font size
plt.xticks(index, classifiers)

# Add value labels above each bar
for i in range(len(classifiers)):
    plt.text(index[i] - bar_width * 1.5, accuracy[i] + 0.005, str(round(accuracy[i], 4)), ha='center', va='bottom')
    plt.text(index[i] - bar_width / 2, precision[i] + 0.005, str(round(precision[i], 4)), ha='center', va='bottom')
    plt.text(index[i] + bar_width / 2, recall[i] + 0.005, str(round(recall[i], 4)), ha='center', va='bottom')
    plt.text(index[i] + bar_width * 1.5, f1_score[i] + 0.005, str(round(f1_score[i], 4)), ha='center', va='bottom')

# Move legend to the middle right side
plt.legend(loc='center right')

# Display the plot
plt.tight_layout()
plt.show()
```
